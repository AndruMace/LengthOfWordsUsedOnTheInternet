```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('unigram_freq.csv', keep_default_na=False)
```

```python
data.head()
```

```python
data['word length'] = data['word'].apply(len)
```

```python
data = data[data['word length'] <= 20]
```

```python
sns.set_style("ticks", {"axes.facecolor": ".1"})
sns.set_context('poster',font_scale=1)
plt.figure(figsize=(12, 16))
mywordplot = sns.stripplot(x="word length", y="count", data=data, jitter=True)
```

```python
figure = mywordplot.get_figure()    
figure.savefig('mywordplot.png', dpi=400)
```
